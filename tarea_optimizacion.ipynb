```julia
using Plots


f(z, x2) = (z - 6*x2) / 2


x2_vals = 0:0.1:5


plt = plot()


niveles = [0, 6, 8, 14, 15, 16]


for z in niveles
    plot!(plt, x2_vals, f.(z, x2_vals), label="Z=$z")
end


display(plt)
```

```julia
using Plots

r1(x2) = (24 - 4x2)/6
r2(x2) = 6 - 2x2
r3(x2) = x2 - 1
r4(x2) = 0

plot(r1, label="6x1+4x2≤24")
plot!(r2, label="x1+2x2≤6")
plot!(r3, label="-x1+x2≤1")
plot!(r4, label="x2≤2")

f(z, x2) = (z - 6*x2) / 2
niveles = [0, 6, 8, 14, 15, 16]

# 🔹 NECESARIO para evaluar la función f con el segundo argumento x2
x2_vals = 0:0.1:5

for z in niveles
    plot!(x2_vals, f.(z, x2_vals), label="Z=$z", linestyle=:dash,color=:darkblue)
end

xlims!(0, 6)
ylims!(-1, 6)


```
